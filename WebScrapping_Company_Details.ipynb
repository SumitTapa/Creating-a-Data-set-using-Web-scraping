```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

final_df = pd.DataFrame()
pages=int(inp
for j in range(1, 503):  # Pages 1 and 2
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'
    }
    url = f"https://www.ambitionbox.com/list-of-companies?page={j}"
    
    try:
        response = requests.get(url, headers=headers, timeout=10)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # Extract company names
        company_names = []
        for company in soup.find_all('h2', class_="companyCardWrapper__companyName"):
            company_names.append(company.text.strip())
        
        # Extract ratings
        ratings = []
        for rating in soup.find_all('div', class_="rating_text"):
            ratings.append(rating.text.strip())
        
        # Extract review counts
        reviews = []
        for review_link in soup.find_all('a', href=True):
            if '/reviews/' in review_link.get('href', ''):
                review_span = review_link.find('span', class_="companyCardWrapper__ActionCount")
                if review_span:
                    reviews.append(review_span.text.strip())
        
        # Extract company niche and HQ
        company_niches = []
        company_hqs = []
        
        for item in soup.find_all('span', class_="companyCardWrapper__interLinking"):
            text_parts = item.text.strip().split('|')
            
            # Extract niche (first part)
            if len(text_parts) > 0:
                company_niches.append(text_parts[0].strip())
            else:
                company_niches.append("N/A")
            
            # Extract HQ (second part)
            if len(text_parts) > 1:
                hq_parts = text_parts[1].strip().split()
                if len(hq_parts) > 1:
                    company_hqs.append(hq_parts[1])
                else:
                    company_hqs.append("N/A")
            else:
                company_hqs.append("N/A")
        
        # Create DataFrame for current page
        df = pd.DataFrame({
            'company_name': company_names,
            'company_niche': company_niches,
            'company_hq': company_hqs,
            'company_rating': ratings,
            'review_count': reviews
        })
        
        # Append to final DataFrame
        final_df = pd.concat([final_df, df], ignore_index=True)
        
        print(f"Page {j}: Found {len(company_names)} companies")
        print(f"  - Ratings: {len(ratings)}")
        print(f"  - Reviews: {len(reviews)}")
        print(f"  - Niches: {len(company_niches)}")
        print(f"  - HQs: {len(company_hqs)}")
        print("-" * 50)
        
    except requests.RequestException as e:
        print(f"Error on page {j}: {e}")
        continue

# Display results
print(f"\n{'='*50}")
print(f"Total companies collected: {len(final_df)}")
print(f"{'='*50}\n")
print(final_df.head(10))

# Optional: Save to CSV
# final_df.to_csv('ambitionbox_companies.csv', index=False)
```
